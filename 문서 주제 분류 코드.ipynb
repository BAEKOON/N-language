```python
import pandas as pd #dataframe
from sklearn.feature_extraction.text import CountVectorizer #DTM 생성
from sklearn.feature_extraction.text import TfidfTransformer #DTM을 TF-IDF 행렬로 변환
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB # 다항분포 나이브 베이즈 모델
```

```python
def file_open(fname) :
#train data 파일 불러오기 및 텍스트 소문자화
    file_str = open(fname).read()
    file = file_str.lower()
    return file
```

```python
trade_doc = []
# trade 파일 단어들 모음
for i in range(1,26) :
    fname = str('./dev/trade/trade_%d.txt'%i)
    doc = file_open(fname)
    trade_doc.append(doc)
print(trade_doc)
```

```python
interest_doc = []
# interest 파일 단어들 모음
for i in range(1,26) :
    fname = str('./dev/interest/interest_%d.txt'%i)
    doc = file_open(fname)
    interest_doc.append(doc)
print(interest_doc)
```

```python
money_supply_doc = []
# money_supply 파일 단어들 모음
for i in range(1,26) :
    fname = str('./dev/money_supply/money_supply_%d.txt'%i)
    doc = file_open(fname)
    money_supply_doc.append(doc)
print(money_supply_doc)
```

```python
jobs_doc = []
# jobs 파일 단어들 모음
for i in range(1,26) :
    fname = str('./dev/jobs/jobs_%d.txt'%i)
    doc = file_open(fname)
    jobs_doc.append(doc)
print(jobs_doc)
```

```python
targetlist = ['interest',]*25 +['jobs',]*25+['money_supply',]*25+['trade',]*25
datalist = interest_doc + jobs_doc + money_supply_doc + trade_doc
df = pd.DataFrame({'data' : datalist ,
                   'target' : targetlist})
#각 문서에 주제를 labeling 할 수 있도록 리스트를 만들어 dataframe 형태로 정리
print(df)
```

```python
dtmvector = CountVectorizer()
tfidf_transformer = TfidfTransformer()
```

```python
x = df.data
y = df.target
from sklearn.model_selection import train_test_split
#dev 파일의 75%는 훈련 데이터, 25%는 모의 테스트 데이터로 나눔
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=2019)
#x_train : 75개 문서 ~ y_train 75개 답
#x_test : 25개 문서 ~ y_test 25개 답
```

```python
train_dtm = dtmvector.fit_transform(x_train)
#훈련용 문서 75개의 등장 단어 1692개로 DTM 생성
print(train_dtm.shape)
```

```python
train_tfidf_vector = tfidf_transformer.fit_transform(train_dtm)
#훈련 데이터 DTM을 tf-idf 형태로 변환
print(train_tfidf_vector.shape)
```

```python
model = MultinomialNB()
#나이브 베이즈 분류기 모델에 적용
model.fit(train_tfidf_vector,y_train)
```

```python
x_test_dtm = dtmvector.transform(x_test) #모의 테스트 데이터를 DTM으로 변환
x_test_tfidfvector = tfidf_transformer.transform(x_test_dtm) #모의 테스트 DTM을 TF-IDF 행렬로 변환
```

```python
from sklearn.metrics import accuracy_score
x_test_predict_value = model.predict(x_test_tfidfvector) #모의 테스트 데이터에 대한 예측
print(accuracy_score(y_test, x_test_predict_value)) #모의 테스트 결과 정확도 출력
```

```python
test_doc = []
#test data 모음
for i in range(1,21) :
    fname = str('./test/%d.txt'%i)
    doc = file_open(fname)
    test_doc.append(doc)
print(test_doc)
```

```python
test_df = pd.DataFrame({'data' : test_doc,})
#test data 리스트를 dataframe 형태로 정리
print(test_df)
```

```python
test_dtm = dtmvector.transform(test_df.data) #테스트 데이터를 DTM으로 변환
test_tfidf_vector = tfidf_transformer.transform(test_dtm) #테스트 데이터 DTM을 TF-IDF 행렬로 변환
```

```python
predict_value = model.predict(test_tfidf_vector) #테스트 데이터에 대한 예측
for i in range(len(predict_value)) : #예측값 출력
    print('%d:'%(i+1),predict_value[i])
```
